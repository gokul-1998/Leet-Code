# Follow up: Could you solve it without converting the integer to a string?

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        # https://youtu.be/OlTk8wM48ww
        
        # step 1 (check if the number is less than 0)
        # check if the number is 0 
        # check if number ends with 0 , as number ending with 0 cannot be an integer
        if x < 0 or (x!=0 and x%10==0):
            return False

        # the variable half will contain the reversed half of the number
        half=0

        # 
        while x>half:
            half = (half*10)+(x%10) # eg x=12321
            x=x//10
            # 1) half = 0+1 = 1 , x=1202
            #  2) half = 10+2 = 12 , x=120
            # 3) half = 120+3 = 123,x = 12
           
        return x==half or x==half//10
        # first one if its symmetric,
        # second one is for non symmetric so we remove the middle number and check if x equals half
            

```

```python
solve=Solution()

print(solve.isPalindrome(121)) # True
print(solve.isPalindrome(-121)) # False
```
