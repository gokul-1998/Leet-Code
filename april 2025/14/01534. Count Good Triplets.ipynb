```python
from typing import List
class Solution:
    def countGoodTriplets(self, arr: List[int], a: int, b: int, c: int) -> int:
        res = 0
        for i in range(0, len(arr)):
            for j in range(i+1, len(arr)):
                for k in range(j+1, len(arr)):
                    if abs(arr[i] - arr[j]) <= a and abs(arr[j] - arr[k]) <= b and abs(arr[k] - arr[i]) <= c: res += 1
                
        return res
```

```python
arr=[3,0,1,1,9,7]
a=7
b=2
c=3


class Solution:
    def countGoodTriplets(self, arr, a, b, c):
        res = 0
        interval = [0] * 1001  # Prefix count array

        for j in range(len(arr)):
            for k in range(j + 1, len(arr)):
                if abs(arr[j] - arr[k]) <= b:
                    left = max(0, max(arr[j] - a, arr[k] - c))
                    right = min(1000, min(arr[j] + a, arr[k] + c))
                    if left <= right:
                        if left == 0:
                            res += interval[right]
                        else:
                            res += interval[right] - interval[left - 1]
            for ind in range(arr[j], 1001):
                interval[ind] += 1

        return res

sol=Solution()
print(sol.countGoodTriplets(arr,a,b,c))
```
