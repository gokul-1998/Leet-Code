# 7. Reverse Integer


## The key to solve this is to convert if the input or the output exceeding the 2**31 criteria to 0

```python
class Solution:
    def reverse(self, x: int) -> int:

        if x >=0 and (x<((2**31) -1)):
            op= int(str(x)[::-1])
            if op>((2**31) -1):
                return 0
            return op
            
        elif x<0 and x > (-1 * (2**31-1)):
            op=  int(str(x*-1)[::-1])*-1
            if op<(-1*(2**31)):
                return 0
            return op
        else:
            return 0
```
